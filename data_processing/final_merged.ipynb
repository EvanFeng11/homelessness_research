此csv后续保存在了"Homelessness_England/analysis/final_merged_cleaned.csv",二者一致。

```python
import pandas as pd

# 读取文件
df_landuse = pd.read_csv('output_data/LAD_landuse_group_stats_wide.csv')
df_socio = pd.read_csv('output_data/merged_census_data.csv')
df_poi = pd.read_csv('output_data/LAD_free_POI_group_with_9_category_summary_wide.csv')

# 统一主键字段名
df_socio = df_socio.rename(columns={'geography code': 'LAD21CD'})

# 合并
df_merged = df_landuse.merge(df_socio, on='LAD21CD').merge(df_poi, on='LAD21CD')

# 查看结果
print(df_merged.head())

#df_merged.to_csv('merged_output_01.csv', index=False)
```

```python
#合并因变量

# 读取已合并的主表
df_main = df_merged

# 读取详细表
df_detail = pd.read_csv('Detailed_LA_20222023_A1.csv')

# 检查详细表主键字段名（假设为"LAD21CD"，如有不同请替换）
# 例如，如果是 'code'，则用 df_detail.rename(columns={'code': 'LAD21CD'}, inplace=True)

# 合并
df_homeless = df_main.merge(df_detail, on='LAD21CD', how='left')

```

```python
#合并crime数据

# 读取主表
df_main2 = df_homeless

# 读取犯罪数据
df_asb = pd.read_csv('output_data/cleaned_crime_asb_sum.csv')
df_total = pd.read_csv('output_data/cleaned_crime_total_sum.csv')

# 重命名主键字段
df_asb = df_asb.rename(columns={'Local Authority code': 'LAD21CD'})
df_total = df_total.rename(columns={'Local Authority code': 'LAD21CD'})

# 合并ASB犯罪数据
df_merge1 = df_main2.merge(df_asb, on='LAD21CD', how='left')

# 合并总犯罪数据
df_final = df_merge1.merge(df_total, on='LAD21CD', how='left')

# 保存结果
df_final.to_csv('output_data/final_merged.csv', index=False)
```
